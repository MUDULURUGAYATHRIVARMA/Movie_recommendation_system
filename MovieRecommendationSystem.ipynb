```python
import pandas as pd
import numpy as ny
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
moviesdata=pd.read_csv('/content/movies.csv')
```

```python
moviesdata.head()
```

```python
moviesdata.shape
```

```python
selected_features=['genres','keywords','tagline','cast','director','overview']
print(selected_features)
```

```python
#fillna() method will replaces null values with specified valus
for feature in selected_features:
  moviesdata[feature]=moviesdata[feature].fillna('')
```

```python
combine_selected_features=moviesdata['genres']+' '+moviesdata['keywords']+' '+moviesdata['tagline']+' '+moviesdata['cast']+' '+moviesdata['director']+' '+moviesdata['overview']
```

```python
print(combine_selected_features)
```

```python
#converting the text to feature vector (numeric values) beacause with help of numeric values we can easily find the cosine similarity
vectorizer=TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combine_selected_features)
print(feature_vector)
```

```python
#to find similarity between user input and dataset we use cosine similarity
similarity=cosine_similarity(feature_vector)
print(similarity)
```

```python
print(similarity.shape)
```

```python
#now getting input from user
moviename=input("Enter your any movie name:")
```

```python
#now taking only list of all movie names to compare with user input
list_movies=moviesdata['title'].tolist()
print(list_movies)
```

```python
# now we are going to find similar or closer to movie tilte with user input
finding_close_movie=difflib.get_close_matches(moviename,list_movies)
print(finding_close_movie)
```

```python
close_match=finding_close_movie[0]
print(close_match)
```

```python
#finding index of movie by using title of movie
index_of_movie=moviesdata[moviesdata.title==close_match]['index'].values[0]
print(index_of_movie)
```

```python
# find similar movies based on index values
similarity_score=list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies acording to their similarities
sorting_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorting_movies)
```

```python
# now we are printing all similar movies using  their index
print("These are few relevent movie_titles according to your search:\n")
i=1
for movie in sorting_movies:
  index = movie[0]
  total_movies=moviesdata[moviesdata.index==index]['title'].values[0]
  if(i<=30):
    print(i,'.',total_movies)
    i+=1


```

Movie Recommendation System


```python
moviename=input("Enter your any movie name:")

list_movies=moviesdata['title'].tolist()

finding_close_movie=difflib.get_close_matches(moviename,list_movies)

close_match=finding_close_movie[0]

index_of_movie=moviesdata[moviesdata.title==close_match]['index'].values[0]

similarity_score=list(enumerate(similarity[index_of_movie]))

sorting_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)

print("\nThese are few relevent movie_titles according to your search:\n")
i=1
for movie in sorting_movies:
  index = movie[0]
  total_movies=moviesdata[moviesdata.index==index]['title'].values[0]
  if(i<=30):
    print(i,'.',total_movies)
    i+=1

```
